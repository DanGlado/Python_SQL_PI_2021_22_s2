# Решение

````python
# 1

crs.execute("SELECT count(STUDENT_ID) FROM EXAM_MARKS WHERE MARK IS NOT NULL AND SUBJ_ID = 73").fetchone()
````

````python
#2 Напишите запрос, который позволяет подсчитать в таблице EXAM_MARKS количество различных предметов обучения.

s = "SELECT COUNT(DISTINCT subj_id) FROM exam_marks"
crs.execute(s).fetchall()
````

````python
# 3 Напишите запрос, который для каждого студента выполняет выборку его идентификатора и минимальной из полученных им оценок.

print(crs.execute("""SELECT STUDENT_ID, min(MARK) 
    FROM EXAM_MARKS 
    GROUP BY STUDENT_ID limit 10""").fetchall())
````

````python
# 4 Напишите запрос, который для каждого студента выполняет выборку его идентификатора и максимальной из полученных им оценок.

s="""
SELECT student_id, max(mark) 
    FROM exam_marks GROUP BY student_id limit 10"""
print(crs.execute(s).fetchall())
````

````python
#5 Напишите запрос, выполняющий вывод первой по алфавиту фамилии студента, начинающейся на букву "И".

s="""
SELECT MIN(SURNAME) FROM STUDENT WHERE SUBSTR(SURNAME, 1,1)='И'
"""
print(crs.execute(s).fetchone())
````

````python
# 5 Var 2

s="""
SELECT MIN(SURNAME) || ' ' || NAME FROM STUDENT WHERE SURNAME LIKE 'И%'
"""
print(crs.execute(s).fetchone())
````

````python
# 5 Var 3
s = """
select surname from student where surname like 'И%' order by surname limit 1;
"""
crs.execute(s).fetchall()
````

````python
# 6. Напишите запрос, который для каждого предмета обучения выводит наименование предмета и 
# максимальное значение номера семестра, в котором этот предмет преподается.

s = """
SELECT subj_name, max(semester) 
    FROM subject GROUP BY subj_name
"""
crs.execute(s).fetchall()
````

````python
#7. Напишите запрос, который для каждого конкретного дня сдачи экзамена выводит данные о количестве студентов, 
# сдававших экзамен в этот день.

s="""
SELECT date(EXAM_DATE), Count(STUDENT_ID) FROM EXAM_MARKS GROUP BY EXAM_DATE
"""
crs.execute(s).fetchall()
````

````python
#8. Напишите запрос, выдающий средний балл для каждого студента.

s="""
SELECT student_id, AVG(mark) FROM exam_marks GROUP BY student_id

"""
print(crs.execute(s).fetchall())
````

````python
#9. Напишите запрос, выдающий средний балл для экзамена по каждому предмету.

s = """
SELECT exam_id, AVG(mark) FROM exam_marks GROUP BY exam_id
"""

print(crs.execute(s).fetchall())
````

````python
# 10. Напишите запрос, определяющий количество сдававших студентов для каждого предмета (оценка больше двух).

s="""
SELECT subj_id, COUNT(student_id)
    FROM exam_marks
    WHERE mark>2
    GROUP BY subj_id;
"""
crs.execute(s).fetchall()
````

````python
# 10. Var 2

s = """
SELECT SUBJECT.SUBJ_NAME, Count(EXAM_MARKS.STUDENT_ID) AS [Количество студентов]
FROM SUBJECT JOIN EXAM_MARKS ON SUBJECT.SUBJ_ID = EXAM_MARKS.SUBJ_ID
where exam_marks.mark > 2
GROUP BY SUBJECT.SUBJ_NAME
"""

crs.execute(s).fetchall()
````

````python
# 16. Напишите запрос, выводящий количество студентов, проживающих в каждом городе. 
# Список отсортировать в порядке убывания количества студентов.

s = """
SELECT CITY, COUNT(STUDENT_ID) AS C 
    FROM STUDENT 
    GROUP BY CITY 
    HAVING CITY IS NOT NULL ORDER BY C DESC
"""
result = crs.execute(s).fetchall()
print(len(result))
result
````

````python
s = """
select university.univ_name, count(student.student_id) as c 
    from university left join student on university.univ_id = student.univ_id 
    group by univ_name order by c desc
"""
result = crs.execute(s).fetchall()
print(len(result))
result
````
